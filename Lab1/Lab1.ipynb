### Ćwiczenie 3

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-6, 6, 100)
x3 = np.linspace(-0, 5, 100)


def fun(x):
    return x**2 + 5

fig, ax = plt.subplots()
plt.plot(x1, fun(x1))
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) = x^2 + 5, x(-1,1)')

fig, ax = plt.subplots()
plt.plot(x2, fun(x2))
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) = x^2 + 5, x(-6,6)')

fig, ax = plt.subplots()
plt.plot(x3, fun(x3))
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) = x^2 + 5, x(0,5)')
```
